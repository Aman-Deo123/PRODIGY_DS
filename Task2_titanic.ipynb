```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the Titanic dataset
titanic_df = pd.read_csv("Titanic-dataset.csv")
```

```python
# Display the first few rows of the dataset
titanic_df.head()
```

```python
# Check for missing values
print(titanic_df.isnull().sum())
```

```python
# Fill missing values in the 'Age' column with the median age
titanic_df['Age'].fillna(titanic_df['Age'].median(), inplace=True)
```

```python
 #Fill missing values in the 'Embarked' column with the most common value
most_common_embarked = titanic_df['Embarked'].mode()[0]
titanic_df['Embarked'].fillna(most_common_embarked, inplace=True)
```

```python
 #Drop 'Cabin' column due to many missing values
titanic_df.drop('Cabin', axis=1, inplace=True)
```

```python
# Explore relationships between variables
# Survival by Passenger Class
sns.countplot(x='Pclass', hue='Survived', data=titanic_df)
plt.title('Survival by Passenger Class') 
plt.show()
```

```python
# Survival by Sex
sns.countplot(x='Sex', hue='Survived', data=titanic_df)
plt.title('Survival by Sex')
plt.show()
```

```python
# Survival by Age
sns.histplot(data=titanic_df, x='Age', hue='Survived', kde=True)
plt.title('Survival by Age')
plt.show()
```

```python
# Survival by Number of Siblings/Spouses
sns.countplot(x='SibSp', hue='Survived', data=titanic_df)
plt.title('Survival by Number of Siblings/Spouses')
plt.show()
```

```python
# Survival by Number of Parents/Children
sns.countplot(x='Parch', hue='Survived', data=titanic_df)
plt.title('Survival by Number of Parents/Children')
plt.show()
```

```python
# Fare distribution
sns.histplot(data=titanic_df, x='Fare', kde=True)
plt.title('Fare Distribution')
plt.show()
```

```python
# Survival by Embarked Port
sns.countplot(x='Embarked', hue='Survived', data=titanic_df)
plt.title('Survival by Embarked Port')
plt.show()
```
